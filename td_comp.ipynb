```python
import pandas as pd

df = pd.read_csv('Data/wiki.csv')
testDF = df[df['category_label'] == 'Games and Toys']
testDF
```

```python
texts = list(testDF['text'])
```

```python
ins = insigen.insigen()

result_dict = {}
for text in texts:
    freq = ins.get_keyword_frequency(text, min_len=2, max_len=3, frequency_threshold=2)
    for key, value in freq.items():
        result_dict[key] = result_dict.get(key, 0) + value

result_dict
```

```python
total_words = result_dict.keys()
total_words
```

```python
testFreq = ins.get_keyword_frequency(df.loc[0]['text'], min_len=2, max_len=3, frequency_threshold=2)
testFreq
```

```python
testFreq = testFreq.keys()
```

```python
def calculate_jaccard_similarity(list1, list2):
    set1 = set(list1)
    set2 = set(list2)
    intersection = len(set1.intersection(set2))
    union = len(set1.union(set2))
    similarity = intersection / union if union != 0 else 0
    return similarity

similarity_percentage = calculate_jaccard_similarity(testFreq, total_words)*100
similarity_percentage
```
